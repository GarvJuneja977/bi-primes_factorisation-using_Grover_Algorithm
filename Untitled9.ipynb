# **Quantum Factorization Setup Using Pennylane**
## This code snippet sets up a quantum computing environment using Pennylane to factorize the number 35. It calculates the number of quantum wires needed based on the logarithm of N, initializes a quantum device for simulation, allocates qubits for storing factorization-related variables (wires_p, wires_q, wires_result), and verifies the wire allocation. This setup is foundational for implementing quantum algorithms, leveraging Pennylane's capabilities for quantum circuit construction and simulation. Adjustments can be made to run on real quantum hardware by specifying appropriate backend providers in Pennylane.

```python
import pennylane as qml
import math

N = 35

# Calculate the number of wires needed
# num_wires is 1 less than the actual numbers required to represent upper_limit to avoid 1 and upper_limit being considered in the solution
num_wires = math.ceil(math.log(N + 1, 2)) - 1

# Initialize the quantum device with the required number of wires
# Using 'default.qubit' for simulation purposes
# If you want to run on a real quantum computer, use a provider like 'qiskit.ibmq'
dev = qml.device('default.qubit', wires=4 * num_wires)
# Example for IBMQ: qml.device('qiskit.ibmq', wires=4*num_wires, backend='ibmq_tokyo', ibmqx_token="XXX")

# Allocate wires for different purposes
# 'wires_p' and 'wires_q' each use num_wires
# 'wires_result' uses 2 * num_wires, hence the total is 4 * num_wires
wires_p = [i for i in range(num_wires)]
wires_q = [i for i in range(num_wires, 2 * num_wires)]
wires_result = [i for i in range(2 * num_wires, 4 * num_wires)]

# Print the wire allocation for verification
print(f"p and q use {num_wires} wires while result uses {2 * num_wires} wires for a total of {4 * num_wires} wires")
```

# **Applying Fourier Transform Phase Shifts in Quantum Circuits.**
## The add_fourier function encapsulates the process of applying phase shifts in a quantum circuit, specifically tailored according to the Fourier transform requirements. It is instrumental in preparing quantum states and implementing quantum algorithms that utilize phase manipulation for computational advantages.


```python
# Function to apply a phase shift according to the Fourier transform
# phase_shift: the phase shift value to apply
# wires: the qubits on which the phase shift is applied
def add_fourier(phase_shift, wires):
    for j in range(len(wires)):
        qml.RZ(phase_shift * np.pi / (2**j), wires=wires[j])
```

# **Quantum Multiplication Using Fourier Transform and Controlled Phase Shifts**
## The multiply function encapsulates the process of performing multiplication using quantum computing techniques. It leverages Quantum Fourier Transform (QFT) for state preparation and applies controlled phase shifts to achieve the multiplication operation on quantum bits (qubits). This function is integral to quantum algorithms aiming to harness quantum parallelism and exploit quantum states for computational tasks beyond classical capabilities.



```python

# Function to perform multiplication using Quantum Fourier Transform (QFT) and controlled phase shifts
# wires_p: the qubits representing the first multiplicand
# wires_q: the qubits representing the second multiplicand
# wires_result: the qubits where the result of the multiplication is stored
def multiply(wires_p, wires_q, wires_result):
    # Apply the Quantum Fourier Transform to the result wires to prepare them for the phase additions
    qml.QFT(wires=wires_result)

    # Iterate over each bit of the first multiplicand (p)
    for i in range(len(wires_p)):
        # Iterate over each bit of the second multiplicand (q)
        for j in range(len(wires_q)):
            # Calculate the coefficient for the phase shift based on the position of the bits
            coeff = 2 ** (len(wires_p) + len(wires_q) - i - j - 2)
            # Apply a controlled phase shift to the result wires, controlled by the current bits of p and q
            qml.ctrl(add_fourier, control=[wires_p[i], wires_q[j]])(coeff, wires_result)

    # Apply the inverse Quantum Fourier Transform to the result wires to return to the computational basis
    qml.adjoint(qml.QFT)(wires=wires_result)
```

# **Quantum Factorization Using Grover's Algorithm in Pennylane**
## The grover_factorization function implements Grover's algorithm for integer factorization using quantum computing principles provided by Pennylane. It begins by applying Hadamard gates to qubits representing potential factors (wires_p and wires_q), creating a superposition state where all states are equally probable. Through a series of iterations (n_epochs), it performs quantum multiplication of the factors and stores intermediate results (wires_result). An oracle flips the sign of states where the product equals N, marking potential solutions. After uncomputing the multiplication using its adjoint operation, a Grover diffusion operator amplifies the probability of correct factor states. Finally, the function measures probabilities of wires_p and wires_q to determine factors of N. This algorithm leverages quantum parallelism and amplitude amplification to potentially solve factorization problems more efficiently than classical algorithms.

```python
# Define the Grover factorization algorithm
# N: the number to be factorized
# wires_p: the qubits representing the first factor
# wires_q: the qubits representing the second factor
# wires_result: the qubits where the intermediate multiplication results are stored
# n_epochs: the number of Grover iterations
@qml.qnode(dev)
def grover_factorization(N, wires_p, wires_q, wires_result, n_epochs):
    # Apply a Hadamard gate to all qubits in wires_p and wires_q to create a superposition state
    for wire in wires_p + wires_q:
        qml.Hadamard(wires=wire)

    # Apply the oracle and Grover search iterations
    for _ in range(n_epochs):
        # Multiply the factors and store the result in wires_result
        multiply(wires_p, wires_q, wires_result)

        # Apply the oracle: flip the sign of the amplitude for the state where the product equals N
        qml.FlipSign(N, wires=wires_result)

        # Uncompute the multiplication by applying the inverse of the multiply function
        qml.adjoint(multiply)(wires_p, wires_q, wires_result)

        # Apply the Grover diffusion operator to amplify the probability of the correct states
        qml.GroverOperator(wires=wires_p + wires_q)

    # Measure the probabilities of the wires in wires_p and wires_q to obtain the factors of N
    return qml.probs(wires=wires_p), qml.probs(wires=wires_q)
```

# **Execution and Measurement**
## It calculates the number of iterations (n_epochs) for Grover's algorithm based on the square root of N, ensuring a sufficient number to maximize the probability of finding factors efficiently. It then executes the grover_factorization function with N, wires_p, wires_q, wires_result, and n_epochs, which implements the quantum factorization process using Pennylane. After running the algorithm, it prints the measurement probabilities (probs_p and probs_q) of potential factors p and q respectively. This approach demonstrates the application of quantum computing principles to factorize integers, showcasing potential factors with their associated probabilities based on quantum measurements.

```python
# Calculate the number of Grover iterations (epochs)
n_epochs = math.ceil(math.sqrt(N))
print(f"The Grover search will run for {n_epochs} epochs")

# Run the Grover factorization algorithm
probs_p, probs_q = grover_factorization(N, wires_p, wires_q, wires_result, n_epochs)

# Print the measurement probabilities of the factors
print("Probabilities of factors p:", probs_p)
print("Probabilities of factors q:", probs_q)
```

# **Probability Distribution of Factors p and q**
## It creates a figure with two side-by-side plots using Matplotlib. Each plot represents the probability distribution of factors p and q obtained from a quantum computation. The first plot (subplot(1, 2, 1)) displays probabilities for factor p, while the second plot (subplot(1, 2, 2)) shows probabilities for factor q. Each bar chart visualizes the likelihood of different quantum states corresponding to potential factors, aiding in understanding the outcomes of the quantum factorization algorithm. Adjustments are made for layout clarity (tight_layout()), and the plots are shown (show()) for visualization.

```python
# Plot the measurement probabilities of the factors
plt.figure(figsize=(12, 6))

# Plot for the first factor (p)
plt.subplot(1, 2, 1)
plt.bar(range(2 ** len(wires_p)), probs_p)
plt.xlabel("Basis states (p)")
plt.ylabel("Probability")
plt.title("Probability distribution of factor p")

# Plot for the second factor (q)
plt.subplot(1, 2, 2)
plt.bar(range(2 ** len(wires_q)), probs_q)
plt.xlabel("Basis states (q)")
plt.ylabel("Probability")
plt.title("Probability distribution of factor q")

# Adjust layout and show the plots
plt.tight_layout()
plt.show()
```

# **Determining Prime Factors from Quantum Computation**
## This code snippet calculates the factors p and q of the bi-prime number N using the probabilities (probs_p) obtained from quantum computation. It converts the probabilities to a numpy array, rounds them to six decimal places, and identifies the index with the highest probability (np.where(result == np.max(result))). Depending on whether one or two indices are found, it determines and prints the prime factors p and q of N. This approach leverages quantum computing to efficiently factorize integers into their prime components.

```python
# Convert the probabilities to an array and round them for comparison
result = np.round(np.array(probs_p), 6)

# Find the index (basis state) with the highest probability
list_result = list(*np.where(result == np.max(result)))

# Determine the factors p and q from the list of results
# If only one result is found, it means both factors are the same (N is a square of a prime)
p, q = list_result if len(list_result) == 2 else (list_result[0], list_result[0])

# Print the factors of the bi-prime number N
print(f"The prime factors of the bi-prime number {N} are {p} and {q}")
```
